#### Invalid order IDs
Question 4: How many invalid order IDs are there in the data frame? An order id is of the form AZ-2011-Y where Y represents a whole number. A Order ID is said to be valid only if Y consists of 7 digits. Find the number of invalid order IDs in the data frame.
    
    a)6
    b)7
    c)8
    d)9


```python
#your code here
s = updated_df["Order ID"]

count = 0
for i,j,k in s.str.split("-"):
#     print(len(k))
    if len(k)!=7:
        count+=1
print(count)
```

```python
updated_df['Order no']=updated_df['Order ID'].str.split('-')

updated_df['Order no']=updated_df['Order no'].apply(lambda x: x[2])

updated_df['no_of_digits']=updated_df['Order no'].apply(lambda x: len(x))

# updated_df['no_of_digits'].value_counts()
updated_df["ID Validation"]=updated_df["no_of_digits"].apply(lambda x: "Valid" if x==7 else "Invalid")
updated_df["ID Validation"].value_counts()
```

#### Column manipulation
Question 7: Remove the orders with negative profit by dropping the corresponding rows with negative `Profit`. Find the product that makes the lowest profit per Quantity in the Technology category.

    a) Nokia Audio Dock, with Caller ID
    b) Logitech Keyboard, Programmable
    c) Motorola Headset, with Caller ID
    d) Belkin Flash Drive, Bluetooth

```python
df_positive_profit=updated_df[(updated_df.Profit>0) & (updated_df.Category=="Technology")]
df_positive_profit["profit_per_quantity"]=df_positive_profit.Profit/df_positive_profit.Quantity
df_positive_profit.sort_values(["profit_per_quantity"]).head(1)
```
